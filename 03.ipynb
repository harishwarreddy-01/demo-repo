```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file directly from the URL
url = "https://raw.githubusercontent.com/opencampus-sh/einfuehrung-in-data-science-und-ml/main/wetter.csv"
df = pd.read_csv(url)

# Drop missing values in the 'Temperatur' column
temperatures = df['Temperatur'].dropna()

# Create a histogram of the 'Temperatur' column
plt.hist(temperatures, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Temperatur (C)')
plt.ylabel('Hufigkeit')
plt.title('Histogramm der Temperatur')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file directly from the URL
url = "https://raw.githubusercontent.com/opencampus-sh/einfuehrung-in-data-science-und-ml/main/wetter.csv"
df = pd.read_csv(url)

# Convert 'Datum' to datetime to extract the month
df['Datum'] = pd.to_datetime(df['Datum'], errors='coerce')

# Group by month and calculate the average temperature
df['Monat'] = df['Datum'].dt.month
monthly_avg = df.groupby('Monat')['Temperatur'].mean()

# Create a bar chart
plt.bar(monthly_avg.index, monthly_avg.values, color='orange', edgecolor='black')
plt.xlabel('Monat')
plt.ylabel('Durchschnittstemperatur (C)')
plt.title('Durchschnittstemperatur pro Monat')
plt.xticks(range(1, 13))
plt.show()
```
